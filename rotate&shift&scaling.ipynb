```python
# Import necessary libraries
from google.colab import files
from skimage import io
from skimage.transform import rotate, resize
import matplotlib.pyplot as plt
import numpy as np

# Upload file
uploaded = files.upload()

# Get the filename of the uploaded file
filename = next(iter(uploaded))

# Read the image
img = io.imread(filename)

# ==========================
# SECTION: Rotate Image
# ==========================
# Rotate the image by 20 degrees
img_rot = rotate(img, 20)

# Ensure the rotated image is in the correct format (uint8)
if img_rot.dtype != np.uint8:
    img_rot = (img_rot * 255).astype(np.uint8)  # Convert float to uint8

# Show the rotated image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_rot)
plt.title("Rotated Image (20 degrees)")
plt.axis('off')

# ==========================
# SECTION: Resize Image
# ==========================
# Resize the image to 100x100 pixels
img_res = resize(img, (100, 100))

# Ensure the resized image is in the correct format (uint8)
if img_res.dtype != np.uint8:
    img_res = (img_res * 255).astype(np.uint8)  # Convert float to uint8

# Show the resized image
plt.subplot(1, 2, 2)
plt.imshow(img_res)
plt.title("Resized Image (100x100)")
plt.axis('off')

# Save the resized image
io.imsave("resized_image.jpg", img_res)

plt.show()
```
