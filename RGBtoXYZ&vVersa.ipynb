```python
# Import necessary libraries
from google.colab import files
from skimage import io, color
import matplotlib.pyplot as plt

# Upload file
uploaded = files.upload()

# Get the filename of the uploaded file
filename = next(iter(uploaded))

# Read the image
img = io.imread(filename)

# Convert to XYZ
img_xyz = color.rgb2xyz(img)

# Convert back to RGB
img_rgb = color.xyz2rgb(img_xyz)

# Show both figures
plt.figure(figsize=(10, 5))

# Display XYZ image
plt.subplot(1, 2, 1)
plt.imshow(img_xyz)
plt.title("Image in XYZ Color Space")
plt.axis('off')  # Optional: to remove axis labels

# Display RGB image (converted back from XYZ)
plt.subplot(1, 2, 2)
plt.imshow(img_rgb)
plt.title("Image Converted Back to RGB")
plt.axis('off')

plt.show()
```
