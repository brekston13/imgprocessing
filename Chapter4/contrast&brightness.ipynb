```python
# Import required packages
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Upload the image
uploaded = files.upload()

# Get the filename of the uploaded image
filename = next(iter(uploaded))

# Read the image using OpenCV
image = cv2.imread(filename)

# Create a dummy image that stores different contrast and brightness values
new_image = np.zeros(image.shape, image.dtype)

# Brightness and contrast parameters
contrast = 3.0
bright = 2

# Change the contrast and brightness of the image
for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
            new_image[y, x, c] = np.clip(contrast * image[y, x, c] + bright, 0, 255)

# Convert images from BGR to RGB for display using matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
new_image_rgb = cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB)

# Display the original and the adjusted images using matplotlib
plt.figure(figsize=(10, 5))

# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')

# Display the adjusted image
plt.subplot(1, 2, 2)
plt.imshow(new_image_rgb)
plt.title("Adjusted Image (Contrast & Brightness)")
plt.axis('off')

plt.show()
```
