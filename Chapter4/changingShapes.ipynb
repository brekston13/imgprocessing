```python
# Import required packages
from google.colab import files
import cv2
import matplotlib.pyplot as plt

# Upload the image
uploaded = files.upload()

# Get the filename of the uploaded image
filename = next(iter(uploaded))

# Read the image
image = cv2.imread(filename)

# ==========================
# SECTION: Erosion Operations
# ==========================

# Define erosion sizes and types
s1, s2, s3 = 0, 10, 10
t1, t2, t3 = cv2.MORPH_RECT, cv2.MORPH_CROSS, cv2.MORPH_ELLIPSE

# Create erosion templates
tmp1 = cv2.getStructuringElement(t1, (2 * s1 + 1, 2 * s1 + 1), (s1, s1))
tmp2 = cv2.getStructuringElement(t2, (2 * s2 + 1, 2 * s2 + 1), (s2, s2))
tmp3 = cv2.getStructuringElement(t3, (2 * s3 + 1, 2 * s3 + 1), (s3, s3))

# Apply erosion to the image
final_erosion1 = cv2.erode(image, tmp1)
final_erosion2 = cv2.erode(image, tmp2)
final_erosion3 = cv2.erode(image, tmp3)

# Convert images from BGR to RGB for display
final_erosion1_rgb = cv2.cvtColor(final_erosion1, cv2.COLOR_BGR2RGB)
final_erosion2_rgb = cv2.cvtColor(final_erosion2, cv2.COLOR_BGR2RGB)
final_erosion3_rgb = cv2.cvtColor(final_erosion3, cv2.COLOR_BGR2RGB)

# Display the eroded images
plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1)
plt.imshow(final_erosion1_rgb)
plt.title("Erosion - RECT")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(final_erosion2_rgb)
plt.title("Erosion - CROSS")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(final_erosion3_rgb)
plt.title("Erosion - ELLIPSE")
plt.axis('off')

# ==========================
# SECTION: Dilation Operations
# ==========================

# Define dilation sizes and types
d1, d2, d3 = 0, 10, 20
t1, t2, t3 = cv2.MORPH_RECT, cv2.MORPH_CROSS, cv2.MORPH_ELLIPSE

# Create dilation templates
tmp1 = cv2.getStructuringElement(t1, (2 * d1 + 1, 2 * d1 + 1), (d1, d1))
tmp2 = cv2.getStructuringElement(t2, (2 * d2 + 1, 2 * d2 + 1), (d2, d2))
tmp3 = cv2.getStructuringElement(t3, (2 * d3 + 1, 2 * d3 + 1), (d3, d3))

# Apply dilation to the image
final_dilation1 = cv2.dilate(image, tmp1)
final_dilation2 = cv2.dilate(image, tmp2)
final_dilation3 = cv2.dilate(image, tmp3)

# Convert images from BGR to RGB for display
final_dilation1_rgb = cv2.cvtColor(final_dilation1, cv2.COLOR_BGR2RGB)
final_dilation2_rgb = cv2.cvtColor(final_dilation2, cv2.COLOR_BGR2RGB)
final_dilation3_rgb = cv2.cvtColor(final_dilation3, cv2.COLOR_BGR2RGB)

# Display the dilated images
plt.subplot(2, 3, 4)
plt.imshow(final_dilation1_rgb)
plt.title("Dilation - RECT")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(final_dilation2_rgb)
plt.title("Dilation - CROSS")
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(final_dilation3_rgb)
plt.title("Dilation - ELLIPSE")
plt.axis('off')

plt.tight_layout()
plt.show()
```
