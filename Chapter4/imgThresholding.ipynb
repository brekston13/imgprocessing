```python
# Import required packages
from google.colab import files
import cv2
import matplotlib.pyplot as plt

# Upload the image
uploaded = files.upload()

# Get the filename of the uploaded image
filename = next(iter(uploaded))

# Read the image
image = cv2.imread(filename)

# ==========================
# SECTION: Thresholding
# ==========================
# Define threshold types
'''
0 - Binary
1 - Binary Inverted
2 - Truncated
3 - Threshold To Zero
4 - Threshold To Zero Inverted
'''

# Apply different thresholds and save in different variables
_, img1 = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY)
_, img2 = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY_INV)
_, img3 = cv2.threshold(image, 50, 255, cv2.THRESH_TRUNC)
_, img4 = cv2.threshold(image, 50, 255, cv2.THRESH_TOZERO)
_, img5 = cv2.threshold(image, 50, 255, cv2.THRESH_TOZERO_INV)

# Convert images from BGR to RGB for display
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3_rgb = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img4_rgb = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
img5_rgb = cv2.cvtColor(img5, cv2.COLOR_BGR2RGB)

# Show the different threshold images
plt.figure(figsize=(12, 10))

plt.subplot(3, 2, 1)
plt.imshow(img1_rgb)
plt.title("Binary Image")
plt.axis('off')

plt.subplot(3, 2, 2)
plt.imshow(img2_rgb)
plt.title("Binary Inverted Image")
plt.axis('off')

plt.subplot(3, 2, 3)
plt.imshow(img3_rgb)
plt.title("Truncated Image")
plt.axis('off')

plt.subplot(3, 2, 4)
plt.imshow(img4_rgb)
plt.title("Threshold to Zero Image")
plt.axis('off')

plt.subplot(3, 2, 5)
plt.imshow(img5_rgb)
plt.title("Threshold to Zero Inverted Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```
