```python
# Import required packages
from google.colab import files
import cv2
import matplotlib.pyplot as plt

# Upload the image
uploaded = files.upload()

# Get the filename of the uploaded image
filename = next(iter(uploaded))

# Read the image for different blurring purposes
image_Original = cv2.imread(filename)
image_MedianBlur = cv2.imread(filename)
image_GaussianBlur = cv2.imread(filename)
image_BilateralBlur = cv2.imread(filename)

# Apply different blurring techniques
image_MedianBlur = cv2.medianBlur(image_MedianBlur, 9)
image_GaussianBlur = cv2.GaussianBlur(image_GaussianBlur, (9, 9), 10)
image_BilateralBlur = cv2.bilateralFilter(image_BilateralBlur, 9, 100, 75)

# Convert images from BGR to RGB for proper display using matplotlib
image_Original_rgb = cv2.cvtColor(image_Original, cv2.COLOR_BGR2RGB)
image_MedianBlur_rgb = cv2.cvtColor(image_MedianBlur, cv2.COLOR_BGR2RGB)
image_GaussianBlur_rgb = cv2.cvtColor(image_GaussianBlur, cv2.COLOR_BGR2RGB)
image_BilateralBlur_rgb = cv2.cvtColor(image_BilateralBlur, cv2.COLOR_BGR2RGB)

# Display the original and blurred images using matplotlib
plt.figure(figsize=(12, 8))

# Display the original image
plt.subplot(2, 2, 1)
plt.imshow(image_Original_rgb)
plt.title("Original Image")
plt.axis('off')

# Display the Median Blurred image
plt.subplot(2, 2, 2)
plt.imshow(image_MedianBlur_rgb)
plt.title("Median Blurred Image")
plt.axis('off')

# Display the Gaussian Blurred image
plt.subplot(2, 2, 3)
plt.imshow(image_GaussianBlur_rgb)
plt.title("Gaussian Blurred Image")
plt.axis('off')

# Display the Bilateral Blurred image
plt.subplot(2, 2, 4)
plt.imshow(image_BilateralBlur_rgb)
plt.title("Bilateral Blurred Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```
