```python
# Import required packages
from google.colab import files
import cv2
import matplotlib.pyplot as plt

# Upload the images
uploaded = files.upload()

# Get the filenames of the uploaded images
filenames = list(uploaded.keys())

# Read image 1 and image 2 using OpenCV
img1 = cv2.imread(filenames[0])
img2 = cv2.imread(filenames[1])

# Resize the second image to match the dimensions of the first image
img2_resized = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Convert the images from BGR to RGB for proper display with matplotlib
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_resized_rgb = cv2.cvtColor(img2_resized, cv2.COLOR_BGR2RGB)

# Define alpha and beta
alpha = 0.30
beta = 0.70

# Blend the images using addWeighted
final_image = cv2.addWeighted(img1, alpha, img2_resized, beta, 0.0)

# Convert the blended image from BGR to RGB for proper display
final_image_rgb = cv2.cvtColor(final_image, cv2.COLOR_BGR2RGB)

# Display the images using matplotlib
plt.figure(figsize=(15, 5))

# Display the first image
plt.subplot(1, 3, 1)
plt.imshow(img1_rgb)
plt.title("Image 1")
plt.axis('off')

# Display the resized second image
plt.subplot(1, 3, 2)
plt.imshow(img2_resized_rgb)
plt.title("Image 2 (Resized)")
plt.axis('off')

# Display the blended image
plt.subplot(1, 3, 3)
plt.imshow(final_image_rgb)
plt.title("Blended Image")
plt.axis('off')

plt.show()
```
