```python
# Import necessary libraries
from google.colab import files
from skimage import io, color
import matplotlib.pyplot as plt

# Upload file
uploaded = files.upload()

# Get the filename of the uploaded file
filename = next(iter(uploaded))

# Read the image
img = io.imread(filename)

# Convert to YIQ
img_yiq = color.rgb2yiq(img)

# Convert back to RGB
img_rgb = color.yiq2rgb(img_yiq)

# Show both figures
plt.figure(figsize=(10, 5))

# Display YIQ image
plt.subplot(1, 2, 1)
plt.imshow(img_yiq)
plt.title("Image in YIQ Color Space")
plt.axis('off')

# Display RGB image (converted back from YIQ)
plt.subplot(1, 2, 2)
plt.imshow(img_rgb)
plt.title("Image Converted Back to RGB")
plt.axis('off')

plt.show()
```
