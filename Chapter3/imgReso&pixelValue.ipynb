```python
# Import necessary libraries
from google.colab import files
from skimage import io
import pandas as pd

# Upload file
uploaded = files.upload()

# Get the filename of the uploaded file
filename = next(iter(uploaded))

# Read the image
img = io.imread(filename)

# Get the resolution (height, width, channels)
resolution = img.shape
print("Image resolution:", resolution)

# Flatten the image pixel values and save to a DataFrame
df = pd.DataFrame(img.flatten(), columns=['Pixel Values'])

# Define maximum rows per sheet
max_rows = 1048576

# Create a Pandas Excel writer using XlsxWriter as the engine
filepath = 'pixel_values.xlsx'
with pd.ExcelWriter(filepath, engine='xlsxwriter') as writer:
    for i in range(0, len(df), max_rows):
        # Slice the DataFrame
        df_slice = df.iloc[i:i + max_rows]
        # Write to a new sheet
        df_slice.to_excel(writer, sheet_name=f'Sheet{i // max_rows + 1}', index=False)

# Download the Excel file
files.download(filepath)
```
