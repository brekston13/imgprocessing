```python
# Import necessary libraries
from google.colab import files
from skimage import io, color
import matplotlib.pyplot as plt

# Upload file
uploaded = files.upload()

# Get the filename of the uploaded file
filename = next(iter(uploaded))

# Read the image
img = io.imread(filename)

# Convert to YPbPr
img_ypbpr = color.rgb2ypbpr(img)

# Convert back to RGB
img_rgb = color.ypbpr2rgb(img_ypbpr)

# Show both figures
plt.figure(figsize=(10, 5))

# Display YPbPr image
plt.subplot(1, 2, 1)
plt.imshow(img_ypbpr)
plt.title("Image in YPbPr Color Space")
plt.axis('off')

# Display RGB image (converted back from YPbPr)
plt.subplot(1, 2, 2)
plt.imshow(img_rgb)
plt.title("Image Converted Back to RGB")
plt.axis('off')

plt.show()
```
