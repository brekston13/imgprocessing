```python
# Import necessary libraries
from google.colab import files
from skimage import io, color
from pylab import *

# Upload file
uploaded = files.upload()

# Read the uploaded image (assuming only one file is uploaded)
image_filename = next(iter(uploaded))
img = io.imread(image_filename)

# Convert to HSV
img_hsv = color.rgb2hsv(img)

# Convert back to RGB
img_rgb = color.hsv2rgb(img_hsv)

# Show both figures
figure(0)
io.imshow(img_hsv)
show()

figure(1)
io.imshow(img_rgb)
show()
```
